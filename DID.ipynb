```python
import pandas as pd
from statsmodels.formula.api import ols

data = {
    'category':    ['Игрушки', 'Игрушки', 'Игрушки', 'Игрушки', 'Подушки', 'Подушки', 'Подушки', 'Подушки'],
    'period':      ['before', 'after', 'before', 'after', 'before', 'after', 'before', 'after'],
    'conv_rate':   [1.5, 1.6, 1.7, 1.8, 1.3, 1.5, 1.4, 1.7],
}
df = pd.DataFrame(data)
```

```python
#Кодируем группу и время. g = 1 для «Игрушки», 0 для «Подушки». t = 1 для «after», 0 для «before»

df['g'] = df['category'].apply(lambda x: 1 if x == 'Игрушки' else 0)
df['t'] = df['period'].apply(lambda x: 1 if x == 'after' else 0)
df['gt'] = df['g'] * df['t']
```

```python
df.head(4)
```

```python
model = ols('conv_rate ~ g + t + gt', data=df).fit()
print(model.summary())
```
